```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
marvel = pd.read_csv('marvel-wikia-data.csv')
marvel['universe'] = 'marvel'
marvel.head()
```

```python
dc = pd.read_csv('dc-wikia-data.csv')
dc['universe'] = 'dc'
dc.head()
```

```python
mar_dc = pd.concat([marvel, dc], sort = False)
mar_dc.head()
```

```python
mar_dc = mar_dc.drop(['page_id', 'urlslug', 'GSM', 'YEAR', 'EYE'], axis = 1)
mar_dc.head()
```

```python
mar_dc = mar_dc.reset_index(drop = True)
mar_dc.head()
```
