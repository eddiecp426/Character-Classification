# Making API Calls for Character Data

```python
powerstats_url = 'https://www.superheroapi.com/api.php/3302070689864554/1/powerstats'
dayjob_url = 'https://superheroapi.com/api/3302070689864554/character-id/work'
relatives_url = 'https://superheroapi.com/api/3302070689864554/character-id/connections'
publisher_url = 'https://superheroapi.com/api/3302070689864554/character-id/biography'
```

```python
results = []
for page in range(0,732):
    page += 1
    url = f'https://www.superheroapi.com/api.php/{api_key}/{page}/powerstats'
    data = requests.get(url)
    results.append(data.content.decode('utf-8'))
```

```python
day_job_results = []
for page in range(0,732):
    page += 1
    url = f'https://superheroapi.com/api/{api_key}/{page}/work'
    data = requests.get(url)
    day_job_results.append(data.content.decode('utf-8'))
```

```python
relatives_results = []
for page in range(0,732):
    page += 1
    url = f'https://superheroapi.com/api/{api_key}/{page}/connections'
    data = requests.get(url)
    relatives_results.append(data.content.decode('utf-8'))
```

```python
publisher_results = []
for page in range(0,732):
    page += 1
    url = f'https://superheroapi.com/api/{api_key}/{page}/biography'
    data = requests.get(url)
    publisher_results.append(data.content.decode('utf-8'))
```

```python
#Creating a list of dictionaries with the key being the feature, and the values being the specific 
#character's stats.
power_list = []
for dict_ in clean_dicts:
    power_dict = {}
    for k,v in dict_.items(): 
        if k == 'name':
            power_dict['name'] = v
        if k == 'intelligence':
            power_dict['intelligence'] = v
        if k == 'strength':
            power_dict['strength'] = v
        if k == 'speed':
            power_dict['speed'] = v
        if k == 'durability':
            power_dict['durability'] = v
        if k == 'power':
            power_dict['power'] = v
        if k == 'combat':
            power_dict['combat'] = v
    power_list.append(power_dict)
power_list[0:5]
```

```python
powerstats_df = pd.DataFrame(power_list)
powerstats_df.head()
```

```python
clean_dj_dicts = []
for result in day_job_results:
    ast.literal_eval(result)
    clean_dj_dicts.append(ast.literal_eval(result))   
clean_dj_dicts[0:5]
```

```python
clean_rel_dicts = []
for result in relatives_results:
    ast.literal_eval(result)
    clean_rel_dicts.append(ast.literal_eval(result))  
clean_rel_dicts[0]
```

```python
clean_pub_dicts = []
for result in publisher_results:
    ast.literal_eval(result)
    clean_pub_dicts.append(ast.literal_eval(result)) 
clean_pub_dicts[0]
```

```python
dj_list = []
for dict_ in clean_dj_dicts:
    dj_dict = {}
    for k,v in dict_.items(): 
        if k == 'id':
            dj_dict['id'] = v
        if k == 'name':
            dj_dict['name'] = v
        if k == 'occupation':
            dj_dict['occupation'] = v
        if k == 'base':
            dj_dict['base'] = v
    dj_list.append(dj_dict)
dj_list[0]
```

```python
rel_list = []
for dict_ in clean_rel_dicts:
    rel_dict = {}
    for k,v in dict_.items(): 
        if k == 'id':
            rel_dict['id'] = v
        if k == 'name':
            rel_dict['name'] = v
        if k == 'relatives':
            rel_dict['relatives'] = v
    rel_list.append(rel_dict)
rel_list[0:2]
```

```python
pub_list = []
for dict_ in clean_pub_dicts:
    pub_dict = {}
    for k,v in dict_.items(): 
        if k == 'id':
            pub_dict['id'] = v
        if k == 'name':
            pub_dict['name'] = v
        if k == 'full-name':
            pub_dict['full-name'] = v
        if k == 'alter-egos':
            pub_dict['alter-egos'] = v
        if k == 'aliases':
            pub_dict['aliases'] = v
        if k == 'publisher':
            pub_dict['publisher'] = v
        if k == 'alignment':
            pub_dict['alignment'] = v
    pub_list.append(pub_dict)
pub_list[0:2]
```

```python
dj_df = pd.DataFrame(dj_list)
dj_df.head()
```

```python
rel_df = pd.DataFrame(rel_list)
rel_df.head()
```

```python
pub_df = pd.DataFrame(pub_list)
pub_df.head()
```

# Merging all DataFrames from API Calls

```python
new_df1 = pd.merge(powerstats_df, dj_df, on = 'name')
new_df1.head()
```

```python
new_df2 = pd.merge(rel_df, pub_df, on = ['name', 'id'])
new_df2.head()
```

```python
char_df = pd.merge(new_df1, new_df2, on = ['id', 'name'])
char_df
```
